```python

import numpy as np
import math
import matplotlib.pyplot as plt

%matplotlib inline
scores = [3.0, 1.0, 0.2]

# softmax for multi classification. Note that the sum of probabilities will be equal to 1 for a given score.
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    x = np.asarray(x)
    exponential = np.exp(x)
    return (exponential/np.sum(exponential,axis=0))
    

print(softmax(scores))

# Plot softmax curves

x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

```python
# when scores are multiplied, they softmax returns more non uniform values
scores = np.array([3.,1.,.2])
scores_10 = scores * 10
print softmax(np.stack([scores,scores_10],axis=1))
plt.plot([1,2,3],softmax(np.stack([scores,scores_10],axis=1)))
```
